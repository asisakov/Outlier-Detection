### Let's proceed to anomaly detection

Probably in this task we will face with point outliers (single data points that lay from the rest distribution), or contextual outliers (punctuations, misprints). In other words it is anomalies. Anomaly Detection (as Outlier Detection) is the part of Unsupervised Learning.

```python
# create the function with hist

def outlierhist(data, column, normalize=False):
    """
    Function for histograms printing
    """
    
    bins = np.linspace(data[column].min(), data[column].max(), 50)
    
    plt.figure(figsize=(16,9))
    
    values = data[column]
    
    if normalize:
        values = np.log10(values + 0.00000001)

    plt.hist(list(values), bins, color='red', label = '$')

    # Plot formatting
    plt.legend()
    plt.xlabel(column)
    plt.ylabel('Number of clients')
    plt.title('Histogram with ' + column)
```

```python
for column in data2.columns:
    outlierhist(data2, column)
```

Now we can set the threshold by hand looking at histograms and find the outlier values. Logic is to prevent spaces between the data. If we have lots of spaces, we try to detect visually the most remote data.

```python
threshold = [60000, 65000, 80, 6200, 1800, 1000, 2000, 63000, 15000, 9.7, 1, 100000, 1, 100000, 100000, 1500]
len(threshold)
```

```python
cols = data2.columns
```

```python
ids = []
for j in range(len(cols)):
    column = cols[j]
    for i in range(len(data2[column])):
        if data2[column][i] > threshold[j]:
            ids.append(i)
            print('client with ID '+ str(i) + ' have an outlier value in column '
                  + column + ' with value = ' + str(data2[column][i]))
```

## 2. Standart deviations is not usable, since the distribution is not normal. But we can try Mean Absolute Deviation, proposed in [3]

```python
mean = data2.ipa_funding.mean()
```

```python
median = data2.ipa_funding.median()
```

```python
median
```

```python
deviations = data2.ipa_funding - median
```

```python
deviations
```

```python
abs_deviations = abs(deviations)
```

```python
mad = abs_deviations.median()
```

```python
mad
```

```python
tol = 6 # set the tolerance - how many deviations are acceptable
norm = 2 # normalization coef for MAD (mean absolute deviation)
```

```python
column = 'ipa_funding'
for i in range(len(data2[column])):
    if data2[column][i] > mean + mad*tol*norm or data2[column][i] < mean - mad*tol*norm:
        print('client with ID '+ str(i) + ' have an outlier value in column '
              + column + ' with value = ' + str(data2[column][i]))
```

```python
plt.figure(figsize=(16,9))
plt.plot(data2.ipa_funding)
plt.plot([mean + mad*tol*norm for i in range(data2.shape[0])])
```

```python
def mad_ids(data, norm=2, tol=7, status=False):
    """
    Function for outliers detection using Mean Absolute Deviation
    """
    mads = []
    means = []
    
    for column in data.columns:
        median = data[column].median()
        means.append(data[column].mean())
        abs_deviations = abs(data2[column] - median)
        mad_local = abs_deviations.median()
        mads.append(mad_local)
    
    cols = data.columns
    ids = []
    for j in range(len(cols)):
        column = cols[j]
        for i in range(len(data[column])):
            if data[column][i] > means[j] + mads[j]*tol*norm:
                ids.append(i)
                if status:
                    print('client with ID '+ str(i) + ' have an outlier value in column '
                          + column + ' with value = ' + str(data2[column][i]))
    return(ids)
        
```

```python
ids = mad_ids(data2, status=True)
```

```python
# ids
```

```python
idsdf = pd.Series(ids)
outliers = idsdf.unique()
print(len(outliers))
outliers.sort()
outliers
```

Here, in contrast to the visualization method, it is challenging to choose normalization coefficients and choose a good tolerance for each case. However, we do not discount this method. It may be suitable for first iteration in finding data emission zones.

```python
def dbscan_visual(data, eps=250, min_samples=2, status=False):
    """
    Outliers visualisation in the data using DBSCAN
    """
    ids = []
    for i in range(data.shape[1]):
        X = data.iloc[:,i:i+1]
        clustering = dbscan(eps=eps, min_samples=min_samples).fit(X)
        print('Outliers number in column', data.columns[i]+':', sum(clustering.labels_ != 0))
        if status:
            plt.figure(figsize=(16,9))
            plt.xlabel('client id')
            plt.ylabel('$')
            plt.title(data.columns[i])
            plt.plot(data[data.columns[i]], label='values')
            plt.plot(data[data.columns[i]]*(clustering.labels_ != 0), label='outliers')
            plt.legend()
        for j in range(len(clustering.labels_)):
            if clustering.labels_[j] < 0:
                ids.append(j)
    
    idsdf = pd.Series(ids)
    outliers = idsdf.unique()
    print('Total number of outliers:', len(outliers))
    outliers.sort()
    return outliers
    
```

```python
ids = dbscan_visual(data2, eps=400, status=True)
```
